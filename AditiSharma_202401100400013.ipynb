```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/student_performance.csv')
```

```python
# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())
```

```python
# Check for missing values in the dataset
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
# Display summary statistics of the dataset
print("\nSummary statistics of numerical columns:")
print(df.describe())
```

```python
# Plot the distribution of final scores
import seaborn as sns # Import seaborn library
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.histplot(df["final_score"], bins=20, kde=True, color='blue')
plt.title("Distribution of Final Scores")
plt.xlabel("Final Score")
plt.ylabel("Count")
plt.show()
```

```python
# Generate a correlation heatmap to see relationships between variables
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Scatter plot: Study Hours vs Final Score to see the impact of study time
plt.figure(figsize=(7, 5))
sns.scatterplot(x=df["study_hours"], y=df["final_score"], color="red")
plt.title("Study Hours vs Final Score")
plt.xlabel("Study Hours")
plt.ylabel("Final Score")
plt.show()
```

```python
# Scatter plot: Sleep Hours vs Final Score to analyze its impact
plt.figure(figsize=(7, 5))
sns.scatterplot(x=df["sleep_hours"], y=df["final_score"], color="green")
plt.title("Sleep Hours vs Final Score")
plt.xlabel("Sleep Hours")
plt.ylabel("Final Score")
plt.show()
```

```python
# Boxplot to analyze the impact of extracurricular activities on final score
plt.figure(figsize=(6, 5))
sns.boxplot(x=df["extracurricular"], y=df["final_score"])
plt.title("Impact of Extracurricular Activities on Final Score")
plt.xlabel("Extracurricular Activities (0 = No, 1 = Yes)")
plt.ylabel("Final Score")
plt.show()
```

```python
# Pairplot to visualize pairwise relationships between numerical features
sns.pairplot(df)
plt.show()
```
